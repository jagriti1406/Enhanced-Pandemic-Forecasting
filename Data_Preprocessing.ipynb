```python
import pandas as pd
import glob

df = df_new.copy()

# List of CSV files
csv_files = glob.glob("/home/jovyan/Pandemic-Forecasting-main/data/Spain/graphs_new/*.csv")

# Iterate over each row in df
for index, row in df.iterrows():
    date = row["date"]
    # Find the CSV file corresponding to the date
    csv_file = f"/home/jovyan/Pandemic-Forecasting-main/data/Spain/graphs_new/ES_{date}.csv"
    
    try:
        mobility_data = pd.read_csv(csv_file)
    except FileNotFoundError:
        print(f"File for {date} not found.")
        continue
    
    # Extract regions and flow data
    confirmed_cases_dict = row["confirmedCases_inflow_outflow_intra"]
    
    # Update each region's inflow, outflow, and intra-region flow
    for region, values in confirmed_cases_dict.items():
        confirmed_cases, inflow, outflow, intra_region_flow = values
        
        # Compute inflow: Sum where `region_2` == region
        inflow = mobility_data.loc[mobility_data['region_2'] == region, 'mobility'].sum()
        
        # Compute outflow: Sum where `region_1` == region
        outflow = mobility_data.loc[mobility_data['region_1'] == region, 'mobility'].sum()
        
        # Compute intra-region flow: Sum where `region_1` == `region_2` == region
        intra_region_flow = mobility_data.loc[
            (mobility_data['region_1'] == region) & (mobility_data['region_2'] == region),
            'mobility'
        ].sum()
        
        # Update the dictionary with the computed values
        confirmed_cases_dict[region] = [confirmed_cases, inflow, outflow, intra_region_flow]
    
    # Update the row in the dataframe
    df.at[index, "confirmedCases_inflow_outflow_intra"] = confirmed_cases_dict
df
```

```python
# Save the updated DataFrame to a new CSV file
df.to_csv("/home/jovyan/Pandemic-Forecasting-main/data/England/england_compact.csv", index=False)
```
