```python
import ast

# List of regions to keep
regions_to_keep = ['madrid', 'barcelona', 'sevilla', 'valencia', 'bizkaia']

df_5r = df.copy()

# Function to filter dictionary for specified regions
def filter_regions(row, regions):
    row = ast.literal_eval(row)
    if isinstance(row, dict):
        return {key: row[key] for key in row if key in regions}
    return row

# Apply the filter to the 'confirmedCases_inflow_outflow_intra' column
df_5r['confirmedCases_inflow_outflow_intra'] = df_5r['confirmedCases_inflow_outflow_intra'].apply(lambda x: filter_regions(x, regions_to_keep))

# Display the filtered DataFrame
display(df_5r)
```

```python
# Initialize dictionaries for each category
confirmed_cases_dict = {}
inflow_dict = {}
outflow_dict = {}
intra_region_dict = {}

start_date = df_new['date'].iloc[0].strftime('%B %d, %Y')  # Directly format the Timestamp
end_date = df_new['date'].iloc[-1].strftime('%B %d, %Y')  # Directly format the Timestamp

# Extract the news column as a list
news_list = df_new['news'].tolist()
policy_list = df_new['policy'].tolist()
discussion_list = df_new['discussion'].tolist()

# Iterate through rows to populate dictionaries
for _, row in df_new.iterrows():
    daily_data = row['confirmedCases_inflow_outflow_intra']
    # # Convert to dictionary
    # daily_data_dict = ast.literal_eval(daily_data)
    for region, values in daily_data.items():
        # Populate lists for each region
        confirmed_cases_dict.setdefault(region, []).append(values[0])
        inflow_dict.setdefault(region, []).append(values[1])
        outflow_dict.setdefault(region, []).append(values[2])
        intra_region_dict.setdefault(region, []).append(values[3])

# Specify the target region
target_region = "barcelona"
country = 'Spain'

# Extract data for the target region
confirmed_cases_region = confirmed_cases_dict.get(target_region, [])
inflow_region = inflow_dict.get(target_region, [])
outflow_region = outflow_dict.get(target_region, [])
intra_region_region = intra_region_dict.get(target_region, [])

# Create the new prompt for the specific region
sample_prompt = f"""
From date {start_date} to date {end_date}, the number of confirmed COVID-19 cases for each day in the region {target_region} in {country} are given in this list {confirmed_cases_region}.
From date {start_date} to date {end_date}, the number of people travelling into the region (i.e. inflow) each day in the region {target_region} in {country} are given in this list {inflow_region}.
From date {start_date} to date {end_date}, the number of people going out of the region (i.e. outflow) each day in the region {target_region} in {country} are given in this list {outflow_region}.
From date {start_date} to date {end_date}, the number of people travelling within the region (i.e. intra region) each day in the region {target_region} in {country} are given in this list {intra_region_region}.
From date {start_date} to date {end_date}, the global news tweets for each day are given in this list {news_list}.
From date {start_date} to date {end_date}, the global policy tweets for each day are given in this list {policy_list}.
From date {start_date} to date {end_date}, the global discussion tweets for each day are given in this list {discussion_list}.
"""

print(sample_prompt)
```

```python
import pandas as pd

# Define the substring where you want to insert
# substring = "Your task is to predict the number of covid-19 cases"
substring = "Question: How many confirmed COVID-19 cases are predicted to occur"
insert_string = " for each day "

# Function to insert the string
def insert_in_prompt(prompt, substring, insert_string):
    if substring in prompt:
        # Find the position of the substring
        index = prompt.find(substring) + len(substring)
        # Insert the small string after the substring
        return prompt[:index] + " " + insert_string + prompt[index:]
    return prompt  # Return original if substring not found

# Apply the function to the column
df['prompt'] = df['prompt'].apply(lambda x: insert_in_prompt(x, substring, insert_string))

# Display the modified DataFrame
display(df)
```
